```python
from langgraph.graph import StateGraph, START, END
from pydantic import BaseModel, Field
from typing import Literal

# Define Pydantic model
class PydanticState(BaseModel):
    name:str
    game:Literal['cricket', 'badminton']

# Define the example node function
def example_node(state: PydanticState):
    return {
        "name": f"{state.name} playing",
        "game": state.game
    }

# Build the state graph
builder = StateGraph(PydanticState)
builder.add_node('example_node', example_node)
builder.add_edge(START, 'example_node')
builder.add_edge('example_node', END)

# Compile the graph
graph = builder.compile()

# Invoke the graph with valid input
result = graph.invoke({"name":"Alice", "game":"badminton"})
print(result) 
```
